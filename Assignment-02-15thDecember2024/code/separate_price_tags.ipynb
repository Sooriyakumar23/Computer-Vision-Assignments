```python
import cv2
import matplotlib.pyplot as plt
```

```python
cv2.__version__
```

```python
price_tags_image_path = "D:\\CV-Krish\\Assignment-02-15thDecember2024\\image_before\\assignment-2.jpg"
```

```python
def readImage(imagePath):
    image = cv2.imread(imagePath)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
price_tags_image = readImage(price_tags_image_path)

height, width, _ = price_tags_image.shape

vertical_padding = height % 6
horizontal_padding = width % 3

top = 0
bottom = 0
left = 0
right = 0

if (vertical_padding > 0):
    vertical_padding = 6 - vertical_padding
    top = vertical_padding // 2
    bottom = vertical_padding - top
if (horizontal_padding > 0):
    horizontal_padding = 3 - horizontal_padding
    left = horizontal_padding // 2
    right = horizontal_padding - left

print ("Top padding =", top)
print ("Bottom padding =", bottom)
print ("Left padding =", left)
print ("Right padding =", right)

```

```python
padded_price_tags_image = cv2.copyMakeBorder(price_tags_image, top, bottom, left, right, cv2.BORDER_REPLICATE)

print ("Padded Tags Image Shape =", padded_price_tags_image.shape)

height_p, width_p, channels = padded_price_tags_image.shape

single_tag_height = height_p // 6
single_tag_width = width_p // 3

plt.imshow(padded_price_tags_image)
plt.show()
```

```python
print("Single Tag Height =", single_tag_height)
print("Single Tag Width =", single_tag_width)
print("Number of Channels =", channels)
```

```python
for i in range(6):
    for j in range(3):
        tag_image = (padded_price_tags_image[
                                            i*single_tag_height:(i+1)*single_tag_height,
                                            j*single_tag_width:(j+1)*single_tag_width
                                        ])
        plt.imshow(tag_image)
        plt.show()
        cv2.imwrite(f"D:\\CV-Krish\\Assignment-02-15thDecember2024\\images_after\\Item_{i*3+j+1}.png", tag_image)
```
